```python
# Graphique - Corrélation entre les notes et le nombre de note
import plotly.express as px

fig = px.scatter(overview_fr, x="averageRating", y='numVotes', color="averageRating",title='Corrélation entre les notes et le nombre de note')
fig.show()
```

```python
#Filtre acteurs et actrices 
actors_principal=overview_fr[['actor1']]#.sort_values('numVotes',ascending=False,na_position ='last').head(5000)
actors_principal.head() 

actors_principal.dropna(subset=["actor1"], axis=0, inplace=True)

actress_principal=overview_fr[['actress1']]#.sort_values('numVotes',ascending=False,na_position ='last').head(5000)
actress_principal.head()
actress_principal.dropna(subset=["actress1"], axis=0, inplace=True)
```

```python
# PB de concatenation


df_actors = pd.DataFrame([[actors_principal]], index = ["Actors_Actress"], columns = ["Actors"])
print(df_actors)
df_actress = pd.DataFrame([[actress_principal]], index = ["Actors_Actress"], columns = ["Actress"])
print(df_actress)


df_gender_acting=np.concatenate([df_actors,df_actress])
df_gender_acting


df_gender_acting
```

```python
#graph Ok
#For Bar charts:

actress_distribution = pd.DataFrame(np.concatenate((overview_fr["actress1"],
                            overview_fr["actress2"],overview_fr["actress3"],
                           overview_fr["actress4"],overview_fr["actress5"]),axis=0)).value_counts()
actress_distribution

#DataFrame Actresses
actress_distribution = pd.DataFrame(actress_distribution).rename_axis("0").reset_index()
actress_distribution
actress_distribution.columns= ["Actress","films_played"]
actress_distribution


#Actors
actor_distribution = pd.DataFrame
actor_distribution = pd.DataFrame(np.concatenate((overview_fr["actor1"],
                            ["actor2"],overview_fr["actor3"],
                            overview_fr["actor4"],overview_fr["actor5"]),axis=0)).value_counts()
actor_distribution
actor_distribution= pd.DataFrame(actor_distribution).rename_axis("0").reset_index()
actor_distribution
#DataFrame Actors
actor_distribution.columns = ["Actor","films_played"]
actor_distribution.head()


```

```python
# Graphique -  Top 10 des actrices
w=actress_distribution.head(10)
fig = px.bar(w,x="Actress",
             y="films_played",
            title='Top 10 des actrices',
             opacity=1,
             width =35,
                         
            )

fig.update_traces(textfont_size=20,
                  marker=dict(color='#CB4154', line=dict(color='#000000', width=4)))


#color='#CB4154
fig.show()
```

```python
# Graphique -  Top 10 des acteurs
w=actor_distribution.head(10)
fig = px.bar(w,x="Actor",
             y="films_played",
            title='Top 10 des acteurs',
            labels = {'quantity of movies played'},
                         
            )

fig.update_traces(textfont_size=20,
                  marker=dict(color='royalblue', line=dict(color='#000000', width=4)))



fig.show()
```
